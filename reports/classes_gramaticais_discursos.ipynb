```
import nltk 
import pandas as pd
from nltk.corpus import stopwords
import gensim
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```
#df = pd.read_csv("./raw_noticias_estadao_2018.csv")
df = pd.read_csv("https://raw.githubusercontent.com/ivynasantino/vies-discursos-deputados/master/data/all_discursos.csv")
df.head()
```

```
nltk.download('stopwords')

# removendo stopwords
default_stopwords = set(stopwords.words('portuguese'))
stopwords_complement = ["nº","", "neste","és", "é", "somos", "sois", "são", "era", "eras", "éreis", "foste", "fostes", "foras",	"fôreis",	"serás", "sereis", "serão",	"serias", "seríeis", "sejas", "sejais", "fosses", "fôsseis", "fores", "fordes", "sê", "sejas", "sejais", "ser", "seres", "sermos",	"serdes",	"serem", "sobre", "é"]

stopwords_complement.extend(default_stopwords)

partidos = ["pp","psdb","cidadania", "avante", "dc", "dem", "mdb", "novo", "patri", "patriota", "pcb", "pcdob", "pco", "phs", "pdt", "pl", "pmb", "pt"]
```

```
def read_questions(row,column_name):
    return gensim.utils.simple_preprocess(str(row[column_name]).encode('utf-8'))
    
documents = []
for index, row in df.iterrows():
  itens = read_questions(row, "discurso")
  lista = []
  for i in itens:
    if len(i) > 3:
      lista.append(i)
    if i in partidos:
      lista.append(i)
    if i not in stopwords_complement:
      lista.append(i)
 
  documents.append(lista)
```

```
words = []
for doc in documents:
  for word in doc:
    words.append(word)
```

```
#text = nltk.word_tokenize(words)
grammar_class = nltk.pos_tag(set(words))
```

```
print(grammar_class)
```

```
adjetivos = ["JJ","JJR","JJS"]
count_adj = 0
ads = []
for cl in grammar_class:
  if cl[1] in adjetivos:
    count_adj += 1
    ads.append(cl)
  
```

```
print(count_adj)
print(ads)
```
